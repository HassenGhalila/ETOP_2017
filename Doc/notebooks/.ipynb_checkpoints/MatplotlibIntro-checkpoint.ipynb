# Comparing Matlab and Python

## Matlab type syntax

Lets start by drawing the function ($y=x^{2}$) with Matlab syntax. First of all, we define the input data by considering a serie of 10 points uniformely spaced in the interval [0,5]. This is done with the instruction <b>linspace</b>. 

```python
x = linspace(0, 5, 10)
y = x ** 2
```

Now, lets plot the curve :

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('Title')
show()
```

Most graphical MATLAB functions exist (or equivalent functions) in <b>pylab</b> modules. For example selecting colors or subplots :

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y, x, 'g*-');
```

The good news is that you can start easily if you are familiar with MATLAB because the similaraty between both syntaxs. Furthermore it has a very simple and intuitive instructions for simple figures. However it is preferable to quickly switch to the Matplotlib API in order to get rapidely familiarised by its Object Oriented tools to create a more complex figures. Indeed, it is very powerfull and easy to use for whom want to create advanced figures with subplots and additional composants.

## Matplotlib object-oriented  API

The main idea with object-oriented programming is to have objects that one can apply functions and actions on, and no object or program states should be global (such as the MATLAB-like API). The real advantage of this approach becomes apparent when more than one figure is created, or when a figure contains more than one subplot.
In order to illustrate this we start like in the previous example, but instead of creating a new global figure instance we store a reference to the newly created figure instance in the **fig** variable, and from it we create a new axis instance axes using the **add_axes** method in the **Figure** class instance **fig** :

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (interval from 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

Even though more code lines were used, the advantage is a full controle of the location and an easiest way to add axis to the figure.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Main Figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Title')

# Inserted Figure
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Inserted title')
```

If we don't want to explicitely define where the axes are placed in the figure, then we can use the many axis layout managers. **subplot** is one of the powerfull tools : 

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title');
```

That was easy, but not so nice especially when axes or labels overlapped (as label y), isn't it ?
To bypass this problem we can use **fig.tight_layout** which ajust automatically positions in a way to avoid overlapping.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')
    
fig.tight_layout()
```

## Size of the figure, aspect ratio and resolution (DPI)
Matplotlib allows you to easily specify the aspect ratio, the DPI and the size of figures when the object <b>Figure</b> is created by passing their values using arguments: <b>figsize</b> and <b>dpi</b>. <b>figsize</b> is a tuple of width and heigh in inches, and <b>dpi</b> is the number of dots per inches. For example, to create a figure with the size 800x400 pixels, we can do :

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

The same arguments may be passed to the layout managers such as for the function **subplots**.

```python
fig, axes = plt.subplots(figsize=(12,3))
x = linspace(0, 5, 10)
y = x ** 2

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Titre');
```

## Legends, labels and titles
After having introduced figures and axes, we can now look at how to annotate graphs with titles, legends, labels, markers, etc. 

** Figure titles **

A title may be added at each axis instance of a figure. To define a title, use the **set_title** method in the axis instance:

```python
ax.set_title("title")
```

** Axes labels **

Similarly, we use the method **set_xlabel** and **set_ylabel** to add labels to X and Y axes:

```python
ax.set_xlabel("x")
ax.set_ylabel("y")
```

**Legends**

Legends may be added in two ways: one is to use the **legend** method of the axis object and  pass a list/tuple of texts associated to the number of plotted curves:

```python
ax.legend(["courbe 1", "courbe 2", "courbe 3"]);
```

The method described above follow the MATLAB API. This method may lead to bugs and errors if curves are deleted or added. A better way is to use the arguments **label="lable text"** in the **plot** method when curves are added and after we use the **legend** method without any argument:

```python
ax.plot(x, x**2, label="courbe 1")
ax.plot(x, x**3, label="courbe 2")
ax.legend()
```

In addition, with this method the legend is automatically updated.
The <b>legend</b> method accepts the argument <b>loc</b> which specify the position of the legen in the figure. The authorised values of <b>loc</b> are indicated by numerical values. See http://matplotlib.org/users/legend_guide.html#legend-location for more details. The most often use are :

```python
ax.legend(loc=0) # Default, let matplotlib decide the right position
ax.legend(loc=1) # Upper right corner
ax.legend(loc=2) # Upper left corner
ax.legend(loc=3) # Bottom left corner
ax.legend(loc=4) # Bottom right corner
# .. many other options are available
```

The following figure gives an example with titles, legends and labels:

```python
fig, ax = subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titre')
ax.legend(loc=2); # coin supérieur gauche
```

## Formating text: LaTeX, font size , font familly
Figure above is usebal as it is but doesn't necessarely correspond to the criteria often requested in the publications. For instance it myght be needed to format the font with *LaTex* tools or to correctly adjust the font size in order to clearly appear in the publications.
In that way, MatlobLib is very interesting because it integrats very well the *LaTex* syntax. For example, we just need to add the *dollar* sign, such as in *LaTex*, to write mathematical formula with the *LaTex* rendering. So we can replace the legend in the above figure by writing **$y=x^3$** (dollar y=x^3 dollar).

Sometime it might be needed to insert the **'r'** character before the legend or labels strings. This problem is due to the fact that the sign (\) is already used in the Python strings and means *escape*. So if you want to write the grec letter **$\alpha$** you will have to replace it by **r'$\alpha$'** or **r"$\alpha$"** (r"dollar \alpha dollar").

```python
fig, ax = subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('Titre')
ax.legend(loc=2); # Upper left corner
```

We can also globaly define the font size and familly which will be applied at all text elements present in figures (Titles, labels, annotations, etc):

```python
# Update all the configuration parameters:
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
```

```python
fig, ax = subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('Title')
ax.legend(loc=2);    # Upper left corner
```

## Setting colors, line widths and line types
### Colors

There are various ways to define colors in a graphics. First of all, we can use the MATLAB syntax where **b** means bleu, **r** means red, etc. The MATLAB API for the line styles are also supported. For instance, **b.-** means bleu line with dot line.

```python
# Line color and style of MATLAB type
ax.plot (x, x ** 2, 'b.-') # Blue line with dots
ax.plot (x, x ** 3, 'g--') # Dashed green line
```

With matplotlib we can also define colors by their name or their hexadecimal value. Also, we can eventually fixe an alpha value, using the key-words arguments **color** and **alpha**: 

```python
fig, ax = subplots()

ax.plot(x, x+1, color="red", alpha=0.5) # Half-transparent red
ax.plot(x, x+2, color="#1155dd") # Hexadecimal RGB code for bleu color
ax.plot(x, x+3, color="#15cc55") # Hexadecimal RGB code for green color
```

### Line styles and markers

To modify the line width, we can use the key-words arguments **linewidth** or **lw**. For the line style we use the key-words **linestyle** or **ls**:

```python
fig, ax = subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# Possible options for line styles: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='-.', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='-.', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='-.', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='-.', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
```

## Controle of the axis appearance

The axes appearance is an important aspect of a figure that we need to frequently change in order to create publication quality graphics. We need to controle the location of the ticks and labels and also modifie the text and size font. We focus on these several parameters in this section.

### Plot range

The first thing we need to set is the axes range. We can do this using the key-words arguments **set_xlim** and **set_ylim** in the axis object. We can also use the **axis('tight')** method to automatically adjust the axes ranges.

```python
fig, axes = subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Adjusted axes ranges")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Custom axes ranges");
```

### Ticks and labels

It is possible to custom the location of the ticks using **set_xticks** and **y_ticks** methods. Both of them generate a list of values specifying their positions along the axes. We can also use the **set_xticklabels** and **set_yticklabels** method that give us a list of text labels associated to each tick position:

```python
fig, ax = subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

It also exists a number of advanced methods that allow to manage major and minor ticks. See http://matplotlib.org/api/ticker_api.html for more details.

### Axes grids

Using the **grid** method in the axis object, we display (or not) grid on the graphics. It is alos possible to costum the apperance of the grid lines using the same key-words arguments previously used with **plot**.

```python
fig, axes = subplots(1, 2, figsize=(10,3))

# Default aspect of the grid
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# Custom grid
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

### Axes properties

It is also possible to modifie the axes properties using the **spines** method:

```python
fig, ax = subplots(figsize=(6,2))

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')

ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

# désactiver l'axe à droite
ax.spines['right'].set_color("none")
ax.yaxis.tick_left() # Ticks and lables only on the left side
```

### Twin axes

Sometimes it is useful to use double X and Y axes. For example, when you plot curves with different units. Matplotlib supports this with **twinx** and **twiny** method:

```python
fig, ax1 = subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"aire $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

### Crossing axes at zero

```python
fig, ax = subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # Set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # Set position of y spine to y=0

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```

##  Color mapping and contour plot

a colormap to encode data. There are a umber of few predefined colormap that you can see in the following site:

```python
alpha = 0.7
phi_ext = 2 * pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * cos(phi_p)*cos(phi_m) - alpha * cos(phi_ext - 2*phi_p)
```

```python
# %load solutions/solution_Mpl2.py
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

# Randomly generate data of different range ...
data1 = np.random.random((10, 10))
data2 = 2 * np.random.random((10, 10))
data3 = 3 * np.random.random((10, 10))

# Configure figures and axes ...
fig, axes = plt.subplots(ncols=3, figsize=plt.figaspect(0.5))
fig.tight_layout()                          # Adjust in a way that subplots fill the figure
cax = fig.add_axes([0.25, 0.1, 0.55, 0.03]) # Add an axis for the colorbar

# You can add your own wishes!
for ax, data in zip(axes, [data1, data2, data3]):
    im = ax.imshow(data, vmin=0, vmax=3, interpolation='nearest')

fig.colorbar(im, cax=cax, orientation='horizontal')
plt.show()
```

## 3D graphics

To use 3D graphics with matplotlib, we first need to create an object of the **Axes3D** class. Then 3D axes can be added to a figure in exactly the same way as done with 2D axes. Also, a convenient manner to do this is to use the key-word argument **projection = '3d'** to the **add_axes** and **add_subplot** method.


```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

### Modify the view angle

Changing the view angle is done using the function **view_init** which accept two arguments: the elevation and the azimuthal angle (in ° units).

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(30, 45)

ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(70, 30)

fig.tight_layout()
```
